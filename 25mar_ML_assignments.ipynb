```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
# df.head()

df[~df['Reviews'].str.isnumeric()]
df_copy = df.copy()
df_copy = df_copy.drop(df_copy.index[10472]) #removed the roe that cantains non numeric values
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
df_copy['Size'].unique()
df_copy['Price'].unique()

df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
df_copy['Size'] = df_copy['Size'].str.replace('k', '')
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
df_copy['Size'] = df_copy['Size'].astype(float)

char_to_remove = ['+',',','$']
cols_to_clean = ['Installs','Price']

for item in char_to_remove:
    for cols in cols_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(item,'')

df_copy['Installs']=df_copy['Installs'].astype(int)
df_copy['Price']=df_copy['Price'].astype(float)

df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Year'] = df_copy['Last Updated'].dt.year

# EDA 
df_copy[df_copy.duplicated('App')].shape
df_copy = df_copy.drop_duplicates(subset='App',keep='first')

numerica_feature = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']  
categorical_feature = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']  

print(f'there are {len(numerica_feature)} numerical  variables which are : {numerica_feature}')
print(f'there are {len(categorical_feature)} categorical variables which are : {categorical_feature}')

plt.figure(figsize=(15,15))
plt.suptitle("Univarate Analysis Of Numerical features" , fontsize=20, fontweight='bold',alpha=0.8, y=1.)
```

## **Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.**

```python
## clean the dataset


## convert installs columns into numeric columns
import pandas as pd

# Load the dataset (assuming you've already loaded it)
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')


# Convert 'Installs' column to numeric format
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = pd.to_numeric(df['Installs'].str.replace('+', ''), errors='coerce')

# Create a frequency table
install_counts = df.groupby('App')['Installs'].sum().reset_index()

# Select the top 10 apps
top_10_apps = install_counts.sort_values(by='Installs', ascending=False).head(10)

# Drop the index column for display
top_10_apps = top_10_apps.reset_index(drop=True)

# Display the top 10 apps and their install counts
top_10_apps
```
