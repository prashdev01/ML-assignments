```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./Datasets/student_data.xlsx')
df.head()
```

```python
df['Result'] = df['Result'].apply(lambda x:1 if x== 'Pass' else 0)
```

```python
df.head()
df['Result']=df['Result'].astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Hours played", y='Result',hue='Hours played', palette='coolwarm')
plt.xlabel("Hours played")
plt.ylabel("Result (0 or 1)")
# plt.ylim(-0.1, 1.1)  # Set y-axis limits to binary values
plt.show()
```
