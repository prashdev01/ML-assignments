```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
df.head()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival Hours']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
```

```python
df['Arrival minutes']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
```

```python
df['Arrival Hours'] = df['Arrival Hours'].astype(int)
df['Arrival minutes'] = df['Arrival minutes'].astype(int)
```

```python
df.head()
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_hours']=df['Dep_Time'].str.split(':').str[0]
df['Dep_mins']=df['Dep_Time'].str.split(':').str[1]
df.drop('Dep_Time',axis=1,inplace=True)
df.head(3)
```

```python
df['Dep_hours']=df['Dep_hours'].astype(int)
df['Dep_mins']=df['Dep_mins'].astype(int)
df.info()
```

```python
df['Duration_hours']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
row_to_delete = df[df['Duration']=='5m'].index[0]
df.drop(row_to_delete,inplace=True)
df['Duration_hours']= df['Duration_hours'].astype(int)
df['Duration_mins'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df['Duration_mins'].fillna(0,inplace=True)
df['Duration_mins'] = df['Duration_mins'].astype(int)
df.drop('Duration',axis=1,inplace=True)
```
